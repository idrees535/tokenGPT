# Manual Fine Tuning

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
model_name = "gpt2"  # or another model variant
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

from torch.utils.data import DataLoader

optimizer = torch.optim.AdamW(model.parameters(), lr=5e-5)
data_loader = DataLoader(your_dataset, batch_size=your_batch_size, shuffle=True)

for epoch in range(num_epochs):
    for batch in data_loader:
        inputs = batch['input_ids'].to(device)
        labels = batch['labels'].to(device)

        outputs = model(inputs, labels=labels)
        loss = outputs.loss

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

model.save_pretrained("your_save_directory")
tokenizer.save_pretrained("your_save_directory")
```
